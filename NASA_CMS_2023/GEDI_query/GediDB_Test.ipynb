```python
import gedidb as gdb
import geopandas as gpd

import random

# Generate a list of 50 random numbers
random_numbers = [random.randint(1, 100) for _ in range(50)]
print(random_numbers)
```

```python
provider = gdb.GEDIProvider(
    storage_type="s3",
    s3_bucket="dog.gedidb.gedi-l2-l4-v002",
    url="https://s3.gfz-potsdam.de"
)

ds = provider.get_data(
    variables=["agbd", "agbd_se", "sensitivity_a2", "elev_lowestmode"],
    geometry=gpd.read_file("/projects/my-public-bucket/Data/NASA_CMS_2023/BRAZIL/GRID/AOI.gpkg"),
    sensitivity=">=0.95",
    sensitivity_a2=">=0.95",
    l2_quality_flag="==1",
    l4_quality_flag="==1",
)

type(ds)
ds.to_dataframe(dim_order=None)
```

```python
df = ds.to_dataframe(dim_order=None)

# Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['longitude'], df['latitude']))

# Set CRS to WGS84 (EPSG:4326)
gdf.set_crs('EPSG:4326', inplace=True)

# Save to GeoPackage file
gdf.to_file("/projects/my-public-bucket/Data/NASA_CMS_2023/BRAZIL/GRID_GEDI/AOI_GEDI.gpkg", layer='points', driver="GPKG")
```
